```python
import cdsapi
from datetime import datetime
import calendar
import numpy as np
```

```python
YEAR = list(np.arange(2023,2024))
MONTH = list(np.arange(1,9))
def DAY_F(year, month):
    DAY = []
    DAY_num = np.arange(1, calendar.monthrange(year, month)[1]+1)
    for day in DAY_num:
        DAY.append(str(day))
    return DAY
```

```python
c = cdsapi.Client()
for year in YEAR:
    for month in MONTH:
        DAY = DAY_F(year, month)
        for day in DAY:
            fileName = 'era5_{}{}{}_{}.nc'.format(str(year),str(month).zfill(2),str(day).zfill(2),'total_precipitation')
            c.retrieve(
            'reanalysis-era5-single-levels',
            {
                'product_type': 'reanalysis',
                'format': 'netcdf',
                'variable': 'total_precipitation',
                'year': str(year),
                'month': [str(month)],
                'day': day,
                'time': [
                    '00:00', '01:00', '02:00',
                    '03:00', '04:00', '05:00',
                    '06:00', '07:00', '08:00',
                    '09:00', '10:00', '11:00',
                    '12:00', '13:00', '14:00',
                    '15:00', '16:00', '17:00',
                    '18:00', '19:00', '20:00',
                    '21:00', '22:00', '23:00',
                ],
                'area': [
                36.7, 90, 24,
                122.1,
                ],
            },
            'D:/ERA5/PRE/' + fileName
            )
```
